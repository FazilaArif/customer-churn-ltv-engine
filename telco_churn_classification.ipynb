# 📞 Project Title

**Telco Customer Churn Prediction Using Machine Learning**

---

## 🧩 Problem Statement

Customer churn is one of the biggest challenges faced by telecommunication companies. Losing existing customers not only reduces revenue but also increases the cost of acquiring new customers.

By analyzing customer demographics, subscription details, billing information, and service usage patterns, businesses can identify customers who are likely to leave their services. This enables companies to take proactive retention measures and improve customer satisfaction.

The objective of this project is to build a machine learning model that accurately predicts whether a customer will **Churn** or **Stay**, allowing businesses to make data-driven decisions.

---

## 📖 Project Description

This project develops a **Customer Churn Prediction System** using supervised machine learning techniques. The model analyzes historical customer information, including demographics, contract details, internet services, billing preferences, tenure, and payment methods, to predict customer churn.

The project covers the complete machine learning workflow, including data preprocessing, exploratory data analysis (EDA), feature encoding, model training, evaluation, and comparison. Multiple classification algorithms are implemented to identify the best-performing model for churn prediction.

The final solution helps telecommunication companies recognize high-risk customers and implement targeted retention strategies to reduce customer attrition.

---

## 📊 Dataset Description

The dataset contains historical records of telecom customers, where each row represents an individual customer along with demographic information, subscribed services, billing details, and churn-related attributes.

### **Feature Description**

* **Gender** – Customer's gender
* **Senior Citizen** – Whether the customer is a senior citizen
* **Partner** – Whether the customer has a partner
* **Dependents** – Whether the customer has dependents
* **Tenure Months** – Number of months the customer has stayed with the company
* **Phone Service** – Availability of phone service
* **Multiple Lines** – Whether multiple phone lines are subscribed
* **Internet Service** – Type of internet service
* **Online Security** – Online security subscription
* **Online Backup** – Online backup service
* **Device Protection** – Device protection plan
* **Tech Support** – Technical support service
* **Contract** – Customer contract type
* **Paperless Billing** – Paperless billing preference
* **Payment Method** – Customer payment method
* **Monthly Charges** – Monthly service charges
* **Total Charges** – Total amount charged
* **Churn Score** – Customer churn risk score
* **CLTV** – Customer Lifetime Value

Each record represents a telecom customer and their subscription details.

## 🎯 Target Variable (Dependent Variable)

### **Churn**

The target variable indicates whether a customer has left the company's services.

* **Yes** – Customer has churned
* **No** – Customer has remained with the company

This is a **Binary Classification** problem where the objective is to predict customer churn.

```python
data_counts = df.groupby(['Senior Citizen', 'Churn']).size().unstack(fill_value=0)

data_percent = data_counts.div(data_counts.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots(figsize=(4,4))
bottom = [0, 0]

for idx, column in enumerate(data_percent.columns):
    ax.bar(data_percent.index, data_percent[column], bottom=bottom, label=column)
    
    for i, val in enumerate(data_percent[column]):
        if val > 0:
            ax.text(i, bottom[i] + val / 2, f'{val:.1f}%', ha='center', va='center', fontsize=10)
    bottom = [i + j for i, j in zip(bottom, data_percent[column])]

ax.set_title("Churn Percentage by Senior Citizen")
ax.set_xlabel("Senior Citizen")
ax.set_ylabel("Percentage (%)")
ax.set_xticks([0, 1])
ax.set_xticklabels(["No", "Yes"])
ax.legend(title="Churn", bbox_to_anchor = (1,1))
sns.despine()
plt.tight_layout()
plt.show()
```

#### 👴 Analyze Churn Percentage by Senior Citizen Status

This visualization compares the **percentage of customers who churned and those who stayed** based on whether they are **Senior Citizens**. Instead of displaying the total number of customers, the chart shows the proportion of each churn category within each group.

A **stacked percentage bar chart** is used to make the comparison fair, even if the number of customers in each group is different. The percentage labels inside the bars make it easy to understand the churn rate for senior and non-senior customers.

This analysis helps determine whether being a senior citizen has an impact on customer churn and provides valuable business insights for designing targeted customer retention strategies.
